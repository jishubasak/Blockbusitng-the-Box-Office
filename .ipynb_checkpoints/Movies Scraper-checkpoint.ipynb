# Movie Details Scraper

```python
ratings = ['PG-13', 'G', 'PG', 'R', 'Not Yet Rated', 'Unrated', 'NC-17']

genres = ['Historical Epic', 'Fantasy', 'Action', 'War', 'Animation', 'Sci-Fi Action',
 'Comedy', 'Sci-Fi Horror', 'Romance', 'Sci-Fi Fantasy', 'Historical Drama',
 'Sci-Fi Adventure', 'Action / Adventure', 'Musical', 'Horror', 'IMAX', 'Romantic Thriller',
 'Foreign', 'Crime Comedy', 'Action Thriller', 'Comedy / Drama', 'Period Drama', 'Romantic Comedy',
 'War Drama', 'Sports Drama', 'Family Adventure', 'Unknown', 'Fantasy Comedy',
 'Horror Comedy', 'Action Drama', 'Family', 'Sports Action', 'Music Drama', 'Drama',
 'Crime Drama', 'Family Comedy', 'Period Adventure', 'Thriller','Sci-Fi Thriller']

compiler = []

try:
    done_list = pd.read_sql('''SELECT * FROM movies''', conn)['link'].to_list()
    df2 = df[~df['link'].isin(done_list)]
    
except:
    print('No table called movies yet.')
    df2 = df

for i in tqdm(range(len(df2))):
    
    movie_title = df2.iloc[i]['title']
    movie_link = df2.iloc[i]['link']
    
    movie_dict = {'title':movie_title, 'link':movie_link, 'domestic_revenue':'',
                 'studio':'', 'release_date':'', 'genres':'', 'runtime':'',
                  'rating':'', 'prod_budget':'', 'worldwide_revenue':'', 
                 'directors':'', 'writers':'', 'actors':'', 'producers':'',
                 'composer':'', 'franchise':0}
    
    url = 'https://boxofficemojo.com' + movie_link
    request = requests.get(url, headers)
    
    if request.status_code == 200:
        
        soup = BeautifulSoup(request.text, 'lxml')
        
        details = soup.select('tr td tr td table tr > td > center > table tr td b')
        
        try:
            movie_dict['domestic_revenue'] = details[0].text
            
            for k in details[1:]:
                try:
                    selector = k.select('a')[0]['href']

                    if 'studio' in selector:
                        movie_dict['studio'] = k.text

                    elif 'date' in selector:
                        movie_dict['release_date'] = k.text

                    else:
                        pass

                except (KeyError, IndexError):

                    if k.text in ratings:
                        movie_dict['rating'] = k.text

                    elif k.text[0] in ['1','2','3','4','5','6','7','8','9','0']:
                        movie_dict['runtime'] = k.text

                    elif k.text[0] == '$':
                        movie_dict['prod_budget'] = k.text

                    elif k.text[:7] == 'Domestic':
                        pass

                    elif k.text in genres:
                        movie_dict['genres'] = k.text

                    else:
                        pass
            
        except IndexError:
            continue
            
  
        try:
            movie_dict['worldwide_revenue'] = soup.select('table tr > td > table:nth-child(2)\
        tr:nth-child(2) > td > table tr > td:nth-child(1) > table tr > td:nth-child(1) > \
        div:nth-child(1) > div.mp_box_content > table tr:nth-child(4) > \
        td:nth-child(2) > b')[0].text
            
        except IndexError:
            movie_dict['worldwide_revenue'] = ''
        
        directors = []
        actors = []
        producers = []
        writers = []
        composers = []
        
        try:
            players = soup.select('table tr > td > table:nth-child(2) tr:nth-child(2) > td > table \
tr > td:nth-child(1) > table tr > td:nth-child(2) > div > div.mp_box_content \
> table tr td font a')
            
            for p in players:
                
                if 'Director' in p['href']:
                    directors.append(p.text)
                
                elif 'Actor' in p['href']:
                    actors.append(p.text)
                
                elif 'Producer' in p['href']:
                    producers.append(p.text)
                
                elif 'Writer' in p['href']:
                    writers.append(p.text)
                    
                elif 'Composer' in p['href']:
                    composers.append(p.text)
                
                else:
                    continue
        except:
            pass
        
        movie_dict['directors'] = ", ".join(directors[1:])
        movie_dict['actors'] = ', '.join(actors[1:])
        movie_dict['writers'] = ', '.join(writers[1:])
        movie_dict['producers'] = ', '.join(producers[1:])
        movie_dict['composer'] = ', '.join(composers[1:])
            
        try:
            franchise = [m.text for m in soup.select('table tr > td > table:nth-child(2) tr:nth-child(2) > td > table tr > td:nth-child(1) > div:nth-child(3) > div.mp_box_content > table tr > td > font > a')]
            
            fran = False
            
            for j in franchise:
                if 'Series' in j:
                    fran = True
            if fran:
                movie_dict['franchise'] = 1
            
        except:
            movie_dict['franchise'] = 0

        compiler.append(movie_dict)
        
        if len(compiler) == 5:
            df_compiler = pd.DataFrame.from_records(compiler)
            df_compiler.to_sql('movies', conn, index=False, if_exists='append')
            compiler = []
            
    else:
        print(f'Status code is {request.status_code} of movie: {movie_title}.')
        continue
    
    print(f'Done at {i}')
    
    sleep(2)
    
```

## Clean up actors dataframe

```python
final = final_df.drop_duplicates().copy()
final.columns = ['actor', 'total_gross', 'num_movies', 'average', 'top_movie', 'gross']
```

```python
final['total_gross'] = final['total_gross'].apply(lambda x: \
                        (float(x.replace('$', '').replace(',','')) * 1000000) if \
                    'k' not in x else (float(x.replace('$', '').replace(',','').replace('k','')) * 1000))
```

```python
final['average'] = final['average'].apply(lambda x: \
                        (float(x.replace('$', '').replace(',','')) * 1000000) if \
                    'k' not in x else (float(x.replace('$', '').replace(',','').replace('k','')) * 1000))
final['gross'] = final['gross'].apply(lambda x: \
                        (float(x.replace('$', '').replace(',','')) * 1000000) if \
                    'k' not in x else (float(x.replace('$', '').replace(',','').replace('k','')) * 1000))
```

```python
final['num_movies'] = final['num_movies'].astype(int)
```

## Assign rankings to actors using quantiles
* column used is `gross`

```python
quantiles = final.quantile([0.25, 0.5, 0.75])
```

```python
final.index = range(len(final))
```

```python
actor_quan = []
for i in range(len(final)):
    if final['gross'][i] <= quantiles.loc[0.25]['gross']:
        actor_quan.append(1)
    elif final['gross'][i] <= quantiles.loc[0.5]['gross']:
        actor_quan.append(2)
    elif final['gross'][i] <= quantiles.loc[0.75]['gross']:
        actor_quan.append(3)
    elif final['gross'][i] > quantiles.loc[0.75]['gross']:
        actor_quan.append(4)
    else:
        pass
```

```python
final['actor_level'] = actor_quan
```
